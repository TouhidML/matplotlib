# Annotation

```python
batter=pd.read_csv('/content/drive/MyDrive/Dataset/Matplot/batter.csv')
batter.sample(5)
```

```python
batter_n=batter.head(100).sample(25)
batter_n
```

```python
plt.figure(figsize=(15,10))
plt.scatter(batter_n['avg'],batter_n['strike_rate'],s=batter_n['runs'])
for i in range(batter_n.shape[0]):
  plt.text(batter_n['avg'].values[i],batter_n['strike_rate'].values[i],batter_n['batter'].values[i])

plt.axhline(120,color='red')
plt.axhline(130,color='green')
plt.axvline(20,color='red')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batter['avg'],batter['runs'])
ax[0,1].scatter(batter['avg'],batter['strike_rate'])
ax[1,0].hist(batter['runs'])
ax[1,1].hist(batter['strike_rate'])
plt.show()
```

# 3D Scatter Plot

```python
batter.sample(5)
```

```python
plt.figure(figsize=(10,10))
ax=plt.subplot(projection='3d')
ax.scatter3D(batter['avg'],batter['runs'],batter['strike_rate'])
ax.set_title('Each batter avg,run and striker_rate')
ax.set_xlabel('avg')
ax.set_ylabel('runs')
ax.set_zlabel('strike_rate')
```

```python
plt.figure(figsize=(10,10))
ax=plt.subplot(projection='3d')
ax.scatter3D(batter['avg'],batter['runs'],batter['strike_rate'])
ax.plot3D(batter['avg'],batter['runs'],batter['strike_rate'],color='red')
ax.set_title('Each batter avg,run and striker_rate')
ax.set_xlabel('avg')
ax.set_ylabel('runs')
ax.set_zlabel('strike_rate')
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2
```

```python
fig=plt.figure(figsize=(10,13))

ax=plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='summer')
plt.colorbar(p)
```

```python
x,y=np.meshgrid(x,y)
z=x**3+y**3
```

```python
fig=plt.figure(figsize=(10,9))

ax=plt.subplot(projection='3d')

p=ax.plot_surface(x,y,z,cmap='winter')
plt.colorbar(p)
```

```python
fig=plt.figure(figsize=(10,9))

ax=plt.subplot(projection='3d')

p=ax.contour(x,y,z,cmap='winter')
plt.colorbar(p)
```

```python
fig=plt.figure(figsize=(10,9))

ax=plt.subplot(projection='3d')

p=ax.contourf(x,y,z,cmap='winter')
plt.colorbar(p)
```
